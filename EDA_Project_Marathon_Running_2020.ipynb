```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the data
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv", low_memory=False)
```

```python
# Preview the data
df.head(10)  # View the first 10 rows
```

```python
df.shape  # Check the shape of the DataFrame
```

```python
df.dtypes  # Inspect data types
```

```python
# Data Cleaning
# Filter for USA races with 50k or 50mi distances in 2020
```

```python
df[df['Event distance/length'] == '50km']
```

```python
# Combining filters into a reusable function for clarity and efficiency
```

```python
def filter_data(data):
    filtered_data = data[
        (data['Event distance/length'].isin(['50km', '50mi'])) &
        (data['Year of event'] == 2020) &
        (data['Event name'].str.contains(r'\(USA\)', na=False))
    ]
    return filtered_data

df_filtered = filter_data(df)
df_filtered.head(10)
```

```python
df_filtered.shape
```

```python
# Remove "USA" from 'Event name' column
```

```python
df_filtered['Event name'].str.split('(').str.get(0)
```

```python
df_filtered['Event name'] = df_filtered['Event name'].str.split('(').str.get(0)
```

```python
df_filtered.head(10)
```

```python
# Clean up athlete age 
```

```python
df_filtered['athlete_age'] = 2020 - df_filtered['Athlete year of birth']
```

```python
# Remove h from athlete perfomance
```

```python
df_filtered['Athlete performance'] = df_filtered['Athlete performance'].str.split(' ').str.get(0)
```

```python
df_filtered.head(5)
```

```python
# Drop columns: Athlete Club, Athlete country, Athlete year of birth, Athlete Age Category
```

```python
df_filtered = df_filtered.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
df_filtered.head()
```

```python
# Clean up NULL values
```

```python
df_filtered.isna().sum()
```

```python
df_filtered[df_filtered['athlete_age'].isna()==1]
```

```python
df_filtered = df_filtered.dropna()
```

```python
df_filtered.shape
```

```python
# Checking for duplicates
df_filtered[df_filtered.duplicated() == True]
```

```python
# Resetting Index 
df_filtered.reset_index(drop = True)
```

```python
# Fixing data types
df_filtered.dtypes.dtypes
```

```python
df_filtered['athlete_age'] = df_filtered['athlete_age'].astype(int)
df_filtered['Athlete average speed'] = df_filtered['Athlete average speed'].astype(float)
```

```python
df_filtered.dtypes
```

```python
df_filtered.head()
```

```python
# Reame columns
```

```python
df_filtered = df_filtered.rename(columns = {'Year of event':'year',
                            'Event dates':'race_day',
                            'Event name':'race_name',
                            'Event distance/length':'race_lenght',
                            'Event number of finishers':'race_number_of_finishers',
                            'Athlete performance':'athlete_performance',              
                            'Athlete gender':'athlete_gender',
                            'Athlete average speed':'athlete_avg_speed',
                            'Athlete ID':'athlete_id',

                            })               
```

```python
df_filtered.head()
```

```python
# Reorder columns
```

```python
df_final = df_filtered[['race_day', 'race_name', 'race_lenght', 'race_number_of_finishers', 'athlete_id', 'athlete_gender', 
           'athlete_age', 'athlete_performance', 'athlete_avg_speed', 'athlete_id']]
```

```python
df_final.head()
```

```python
df_final.dtypes
```

```python
# Categorizing athletes into age groups and ensure numeric columns are ready for analysis.
```

```python
# Create age groups

bins = [0, 18, 30, 40, 50, 60, 100]
labels = ['<18', '18-30', '30-40', '40-50', '50-60', '60+']
df_final['age_group'] = pd.cut(df_final['athlete_age'], bins=bins, labels=labels, right=False)

# Check the distribution of age groups
print(df_final['age_group'].value_counts())
```

```python
#Creating Boxplot: Average Speed by Gender and Age Group

#This visualization will show how average speed varies across gender and age groups
```

```python
# Boxplot of average speed by gender and age group

plt.figure(figsize=(12, 6))
sns.boxplot(data=df_final, x='age_group', y='athlete_avg_speed', hue='athlete_gender', palette='coolwarm')
plt.title('Athlete Average Speed by Gender and Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Speed')
plt.legend(title='Gender')
plt.show()
```

```python
# Creating Bar Chart: Top Races by Number of Finishers

# We’ll highlight the top 10 races with the most finishers
```

```python
# Get top 10 races by number of finishers

top_races = df_final.groupby('race_name')['race_number_of_finishers'].sum().sort_values(ascending=False).head(10).reset_index()

# Bar chart

plt.figure(figsize=(10, 6))
sns.barplot(data=top_races, x='race_number_of_finishers', y='race_name', palette='viridis')
plt.title('Top 10 Races by Number of Finishers')
plt.xlabel('Number of Finishers')
plt.ylabel('Race Name')
plt.show()
```

```python
# Creating Heatmap: Correlation Between Numeric Variables

# Visualize relationships between numeric columns like age, speed, and performance.
```

```python
# Correlation heatmap

plt.figure(figsize=(8, 6))
numeric_cols = ['athlete_age', 'athlete_avg_speed', 'race_number_of_finishers']
corr_matrix = df_final[numeric_cols].corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Numeric Variables')
plt.show()
```

```python
# Creating Line Plot: Trends in Finishers Over Time

# We’ll analyze how the number of finishers has changed over time
```

```python
# Group data by race_day

time_trend = df_final.groupby('race_day')['race_number_of_finishers'].sum().reset_index()

# Line plot

plt.figure(figsize=(16, 8))  # Increase figure size
sns.lineplot(data=time_trend, x='race_day', y='race_number_of_finishers', marker='o')

plt.title('Number of Finishers Over Time')
plt.xlabel('Race Day')
plt.ylabel('Number of Finishers')

plt.xticks(rotation=90)  # Rotate labels to 90 degrees
plt.locator_params(axis='x', nbins=10)  # Show fewer x-axis ticks (adjust nbins as needed)

plt.show()
```

```python
# Save cleaned dataset to a new CSV file
def save_cleaned_data(data, output_path):
    data.to_csv(output_path, index=False)
    print(f"Cleaned dataset saved to {output_path}")

output_file_path = "Final_UM_Races.csv"
save_cleaned_data(df_filtered, output_file_path)
```
